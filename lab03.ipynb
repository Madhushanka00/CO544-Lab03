```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def posteriorPlot(nx, ny, m1, C1, m2, C2, P1, P2):
  x = np.linspace(-5, 5, nx)
  y = np.linspace(-5, 5, ny)
  X, Y = np.meshgrid(x, y, indexing='ij')

  Z = np.zeros([nx, ny])
  for i in range(nx):
    for j in range(ny):
      xvec = np.array([X[i,j], Y[i,j]])
      num = P1*gauss2D(xvec, m1, C1)
      den = P1*gauss2D(xvec, m1, C1) + P2*gauss2D(xvec, m2, C2)
      Z[i,j] = num / den
  return X, Y, Z
```

```python
def gauss2D(x, m, C):
    Ci = np.linalg.inv(C)
    dC = np.linalg.det(C)
    num = np.exp(-0.5 * np.dot((x-m).T, np.dot(Ci, (x-m))))
    den = 2 * np.pi * np.sqrt(dC)
    return num / den

def twoDGaussianPlot(nx, ny, m, C):
    x = np.linspace(-5, 10, nx)
    y = np.linspace(-5, 10, ny)
    X, Y = np.meshgrid(x, y, indexing='ij')
    Z = np.zeros([nx, ny])
    for i in range(nx):
        for j in range(ny):
            xvec = np.array([X[i,j], Y[i,j]])
            Z[i,j] = gauss2D(xvec, m, C)
    return X, Y, Z
```

```python
def plot_contours(x, y, z, levels=20):
    """
    Plots contours of a 2D function.

    Args:
        x: 1D array of x-axis values.
        y: 1D array of y-axis values.
        z: 2D array of function values.
        levels: Number of contour levels to plot.
    """

    # Create the plot
    fig, ax = plt.subplots()

    # Plot the contours
    contours = ax.contour(x, y, z, levels=levels)

    # Add labels and title
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Contour Plot")

    # Show the plot
    plt.show()
```

```python
m1=np.array([0,3])

m2=np.array([3,2.5])

C1=C2=np.array([[2, 1], [1, 2]], np.float32)

P1=P2=0.5

nx , ny = 50,40

x1,y1,z1 = twoDGaussianPlot(nx, ny, m1, C1)
x2,y2,z2 = twoDGaussianPlot(nx, ny, m2, C2)

NumDataPerClass = 200
A = np.linalg.cholesky(C1)
U1 = np.random.randn(NumDataPerClass,2)
X1 = U1 @ A.T + m1
U2 = np.random.randn(NumDataPerClass,2)
X2 = U2 @ A.T + m2

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X1[:,0], X1[:,1], c="c", s=10)
ax.scatter(X2[:,0], X2[:,1], c="m", s=10)
ax.set_xlim(-5, 10)
ax.set_ylim(-5, 10)
plt.contour(x1,y1,z1, aspect='equal')

plt.contour(x2,y2,z2, aspect='equal')
plt.show()

x3,y3,z3 = posteriorPlot(nx, ny, m1, C1, m2, C2, P1, P2)
plt.contour(x3,y3,z3,6)
```

```python
import numpy as np

# Assuming m1, m2, C1, C2 are defined
# Compute the Fisher Linear Discriminant direction
wF = np.linalg.inv(C1 + C2) @ (m1 - m2)

proj1 = np.dot(X1, wF)
proj2 = np.dot(X2, wF)

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot the projected data
ax.hist(proj1, alpha=0.5, label='X 1')
ax.hist(proj2, alpha=0.5, label='X 2')

# Add a legend
ax.legend(loc='upper right')

# Show the plot
plt.show()
```

4. Project the data onto the Fisher discriminant directions and plot histograms of the distribution of projections (an example of this is in Fig. 2(a);

```python
import matplotlib


Ci = np.linalg.inv(2*C1)
uF = Ci @ (m2-m1)
yp1 = X1 @ uF
yp2 = X2 @ uF
matplotlib.rcParams.update({'font.size': 16})
plt.hist(yp1, bins=40)
plt.hist(yp2, bins=40)
plt.savefig('histogramprojections.png')
```

```python
# Define a range over which to slide a threshold
#
pmin = np.min( np.array( (np.min(yp1), np.min(yp2) )))
pmax = np.max( np.array( (np.max(yp1), np.max(yp2) )))
print(pmin, pmax)
# Set up an array of thresholds
#
nRocPoints = 50;
thRange = np.linspace(pmin, pmax, nRocPoints)
ROC = np.zeros( (nRocPoints, 2) )
# Compute True Positives and False positives at each threshold
#
for i in range(len(thRange)):
    thresh = thRange[i]
    TP = len(yp2[yp2 > thresh]) * 100 / len(yp2)
    FP = len(yp1[yp1 > thresh]) * 100 / len(yp1)
    ROC[i,:] = [TP, FP]
# Plot ROC curve
#
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(ROC[:,1], ROC[:,0], c='m')
ax.set_xlabel('False Positive')
ax.set_ylabel('True Positive')
ax.set_title('Receiver Operating Characteristics')
ax.grid(True)
plt.savefig('rocCurve.png')
```

```python
from numpy import trapz

# Compute the area under the ROC curve
auc = trapz(ROC[:, 0], x=ROC[:, 1])

print('AUC-ROC:', auc)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters
mean_A = [0, 3]
mean_B = [3, 2.5]
# C1=C2=np.array([[2, 1], [1, 2]], np.float32)
C1 = np.array([[2, 0], [0, 2]], np.float32)
C2 = np.array([[4, 0], [0, 4]], np.float32)
num_samples = 200

# Generate synthetic data
np.random.seed(0)
data_A = np.random.multivariate_normal(mean_A, C1, num_samples)
data_B = np.random.multivariate_normal(mean_B, C2, num_samples)

# Plot the data
plt.scatter(data_A[:, 0], data_A[:, 1], label='Class A')
plt.scatter(data_B[:, 0], data_B[:, 1], label='Class B')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Synthetic Data for Classification')
plt.show()
```

```python
# Distance-to-Mean Classifier
def distance_to_mean_classifier(point, mean_A, mean_B):
    dist_A = np.linalg.norm(point - mean_A)
    dist_B = np.linalg.norm(point - mean_B)
    return 0 if dist_A < dist_B else 1

# Mahalanobis Distance-to-Mean Classifier
cov_invA = np.linalg.inv(C1)
cov_invB = np.linalg.inv(C2)
def mahalanobis_distance_classifier(point, mean_A, mean_B):
    mahalanobis_A = np.sqrt((point - mean_A).T @ C1 @ (point - mean_A))
    mahalanobis_B = np.sqrt((point - mean_B).T @ C2 @ (point - mean_B))
    return 0 if mahalanobis_A < mahalanobis_B else 1
```

```python
# Visualize decision boundaries
x_min, x_max = -4, 7
y_min, y_max = -4, 7
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 200))
mesh_points = np.c_[xx.ravel(), yy.ravel()]
```

```python
predictions_distance = [distance_to_mean_classifier(point, mean_A, mean_B) for point in mesh_points]
predictions_mahalanobis = [mahalanobis_distance_classifier(point, mean_A, mean_B) for point in mesh_points]
```
